## BPE Example Algo from [Paper](https://arxiv.org/abs/1508.07909)

```python
import re, collections
```

```python
vocab = {'l o w </w>':5, 'l o w e r </w>':2, 'n e w e s t </w>':6, 'w i d e s t </w>':3}
```

```python
def get_stats(vocab):
    # defaultdict int makes sure that if someone access a non-existent element, it would return a default 0 instead of error.
    pairs = collections.defaultdict(int)
    for word, freq in vocab.items():
        symbols = word.split()
        for i in range(len(symbols)-1):
            pairs[symbols[i], symbols[i+1]] += freq
    return pairs
```

```python
for word, freq in vocab.items():
    print("*"*100)
    print(f"{word}: {freq}")
    symbols = word.split()
    print(f"Word: {word}\tSymbols: {symbols}")
    for i in range(len(symbols)-1):
        # print(f"{i}: {symbols[i]}")
        pairs[symbols[i], symbols[i+1]] += freq
    print(f"Pairs: {pairs}")
```

```python
def merge_vocab(pair, v_in):
    v_out = {}
    bigram = re.escape(' '.join(pair))
    p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
    for word in v_in:
        w_out = p.sub(''.join(pair), word)
        v_out[w_out] = v_in[word]
    return v_out
```

```python
num_merges = 10 # Hyperparam of the algo.
```

```python
for i in range(num_merges):
    print("*"*50)
    pairs = get_stats(vocab)
    print(f"Vocab: {vocab}\nPairs: {pairs}")
    best = max(pairs, key=pairs.get)
    print(f"Best: {best}")
    break
```
